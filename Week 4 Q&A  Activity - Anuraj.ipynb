```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("D:\DATA SCIENCE\TASKS\Telco_Customer_Churn.csv")
data
```

# 1.Comparing churn count with respect to gender

```python
# 1.Comparing churn count with respect to gender
data['gender'].count()
plt.figure(figsize=(5,5))
sns.catplot(x='Churn',data=data,hue='gender', kind='count', )
plt.title('Genderwise Churn Count',fontsize=15)
```

# 2 Findout how many female senior citizen there in the dataset

```python
data.groupby('gender').sum()
```

# 3 Compare 'tenure' with 'Total Charges'

```python
import matplotlib.pyplot as plt
plt.figure(figsize =(10,10))
sns.scatterplot(data=data, x="tenure",y="TotalCharges", color='green', s=10).set(title='Total charges vs Tenure scatter plot')
```

# 4.Find out which contract is preferred by the senior citizen.

```python
#4.Find out which contract is preferred by the senior citizen.
sen_cit=data.copy()
sen_cit.loc[sen_cit['SeniorCitizen']==0] ## filters out senior citizens
sns.catplot(x='Contract', kind="count", data=sen_cit).set(title='Senior citizens contract preference')
```

# 5 Comment your finds on Payment Method

```python
data['PaymentMethod'].value_counts(normalize=True).plot.pie(autopct='%.1f %%', ylabel='', legend=False, title='Popularity of various payment methods')
```

```python
plt.figure(figsize=(12,12))
sns.catplot(x='PaymentMethod', hue='gender', kind='count', 
            data=data).set(title='Genderwise Payment method preference').set_xticklabels(rotation=90)
```
