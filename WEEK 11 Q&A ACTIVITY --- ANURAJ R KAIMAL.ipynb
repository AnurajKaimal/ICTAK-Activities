```python
data['passed']=data['passed'].map({'yes':1,'no':0})
```

```python
X=data.drop(['passed'],axis=1)
y=data['passed']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_en = LabelEncoder()
l1 = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'schoolsup', 'famsup',
       'paid', 'activities', 'Mjob','Fjob', 'nursery', 'higher', 'internet', 'romantic']
for x in l1:
    data[x] = lab_en.fit_transform(data[x])
```

```python
data.head()
```

```python
data_obj=[f for f in X.columns if X[f].dtypes=='O']
X=pd.get_dummies(X,columns=data_obj,drop_first=True)
X.head()
```

#### SPLITTING TO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=40,test_size=.35) #35% data is for testing
```

### LOGISTIC REGERESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
log_model=LogisticRegression(solver='newton-cg')
log_model.fit(X_train,y_train)
prediction=log_model.predict(X_test)
print("F1_score :",f1_score(y_test,prediction))
print("Accuracy score :",accuracy_score(y_test,prediction))
print("Confusion_matrix :",confusion_matrix(y_test,prediction))
```

### RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rf_mod = RandomForestClassifier(random_state=40)
rf_mod.fit(X_train,y_train)
y_pred=rf_mod.predict(X_test)
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print('Conclusion matrix\n',confusion_matrix(y_test,y_pred))
```

### SVM MODEL

```python
from sklearn.svm import SVC
svm_gauss=SVC(kernel='rbf')
svm_gauss.fit(X_train,y_train)
y_pred=svm_gauss.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print('Conclusion matrix\n',confusion_matrix(y_test,y_pred))
```
