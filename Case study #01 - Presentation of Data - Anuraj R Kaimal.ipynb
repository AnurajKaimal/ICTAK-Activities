# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
import numpy as np 
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Gender']
Females = [5052]
Males = [4948]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.1, Females, 0.1, label = 'Female')
plt.bar(X_axis + 0.1, Males, 0.1, label = 'Male')
  
plt.xticks(X_axis, X)
plt.ylabel("Number of Cars Bought")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```
