```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("D:\DATA SCIENCE\TASKS\shipdata.csv")
data
```

```python
data.set_index('PassengerId',inplace=True)
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.Embarked.fillna(data.Embarked.mode()[0], inplace = True)
```

```python
data.Cabin = data.Cabin.fillna('Unknown Cabin')
```

```python
data['Salutation'] = data.Name.apply(lambda name: name.split(',')[1].split('.')[0].strip())
```

```python
grp = data.groupby(['Sex', 'Pclass'])
```

```python
grp.Age.apply(lambda x: x.fillna(x.median()))
data.Age.fillna(data.Age.median, inplace = True)
```

```python
sns.boxplot(data['Fare'],data=data)
```

```python
Q1 = data['Fare'].quantile(0.25)
Q3 = data['Fare'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
data['Fare']=np.where(data['Fare']>upper_whisker,upper_whisker,np.where(data['Fare']<lower_whisker,lower_whisker,data['Fare']))
```
